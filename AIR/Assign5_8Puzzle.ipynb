```python
class Node:
    def __init__(self, data, level, fval):
        self.data = data
        self.level = level
        self.fval = fval
    
    
    def find(self, puz_mat, search):
        for i in range(len(self.data)):
            for j in range(len(self.data)):
                if(puz_mat[i][j] == search):
                    return i, j
    
    
    def copy(self, puz_mat):
        temp_puz_mat = []
        for i in puz_mat:
            temp = []
            for j in i:
                temp.append(j)
            temp_puz_mat.append(temp)
        return temp_puz_mat
    
    
    def shuffle(self, puz_mat, x1, y1, x2, y2):
            if(x2 >= 0 and x2 < len(self.data) and y2 >= 0 and y2 < len(self.data)):
                temp_puz_mat = self.copy(puz_mat)
                temp_puz_mat[x1][y1], temp_puz_mat[x2][y2] = temp_puz_mat[x2][y2], temp_puz_mat[x1][y1]
                return temp_puz_mat
            return None
    
    
    def generate_child(self):
        x, y = self.find(self.data, 0)
        val_list = [[x, y-1], [x, y+1], [x-1, y], [x+1, y]]
        children = []
        for i in val_list:
            child = self.shuffle(self.data, x, y, i[0], i[1])
            if child is not None:
                child_node = Node(child, self.level+1, 0)
                children.append(child_node)
        return children
```

```python
class Puzzle:
    def __init__(self, size):
        self.n = size
        self.open = []
        self.closed = []
        
    
    def accept(self):
        puz = []
        for i in range(self.n):
            temp = list(map(int, input().split(" ")))
            puz.append(temp)
        return puz
    
    
    def f(self, start, goal):
        return self.h(start.data, goal) + start.level
    
    
    def h(self, start, goal):
        temp = 0
        for i in range(self.n):
            for j in range(self.n):
                if start[i][j] != goal[i][j] and start[i][j] != 0:
                    temp += 1
        return temp
    
    
    def print_current_state(self, state):
        for i in state.data:
            for j in i:
                print(j, end = " ")
            print("")
        print("\n")
    
    
    def process(self):
        print("Enter start state matrix\n")
        start = self.accept()
        
        print("\nEnter goal state matrix\n")
        goal = self.accept()
        
        start = Node(start, 0, 0)
        start.fval = self.f(start, goal)
        
        self.open.append(start)
        
        print("\n\nMoves\n")
        
        while True:
            current = self.open[0]
            self.print_current_state(current)
            
            if( self.h(current.data, goal) == 0):
                break
            
            for i in current.generate_child():
                i.fval = self.f(i, goal)
                self.open.append(i)
            self.closed.append(current)
            del self.open[0]
            
            self.open.sort(key = lambda x:x.fval, reverse=False)
```

```python
if __name__ == '__main__':
    print("Enter size of puzzle")
    inp = int(input())
    puzz = Puzzle(inp)
    puzz.process()
```

```python
def check_safe(row, col, occupied_rows, occupied_pos_diagonal, occupied_neg_diagonal):
    #if row or diagonal is occupied return false
    #No need to check column
    if(occupied_pos_diagonal[row + col] or occupied_neg_diagonal[row - col + N - 1] or occupied_rows[row]):
        return False
    return True

def solveNQueens(board, N, col, occupied_rows, occupied_pos_diagonal, occupied_neg_diagonal):
    if col >= N:
        #Return true as all Queens are placed
        return True
    
    for row in range(N):
        #check in which row queen can be placed
        if check_safe(row, col, occupied_rows, occupied_pos_diagonal, occupied_neg_diagonal):
            
            #Safe to place Queen
            board[row][col] = 1
            
            #Mark Diagonals as occupied
            occupied_pos_diagonal[row + col] = True
            occupied_neg_diagonal[row - col + N - 1] = True
            
            #Mark row as occupied
            occupied_rows[row] = True
            
            #Try to place rest of the Queens using recursion
            if solveNQueens(board, N, col+1, occupied_rows, occupied_pos_diagonal, occupied_neg_diagonal):
                return True
            
            #All Queens Cannot be placed
            #Mark current place as not occupied
            board[row][col] = 0
            
            #Mark diagonals as not occupied
            occupied_pos_diagonal[row + col] = False
            occupied_neg_diagonal[row - col + N - 1] = False
            
            #Mark row as not occupied
            occupied_rows[row] = False
            
    #All rows are occupied
    return False

def solve(N):
    #Generate Board of size N*N
    board = [[0 for i in range(N)] for j in range(N)]
    
    #List for occupied rows
    occupied_rows = [False] * N
    
    #List for occupied diagonals
    number_of_diagonals = (2 * N) - 1
    occupied_pos_diagonal = [False] * number_of_diagonals
    occupied_neg_diagonal = [False] * number_of_diagonals
    
    #Solve the problem
    if solveNQueens(board, N, 0, occupied_rows, occupied_pos_diagonal, occupied_neg_diagonal):
        return True, board
    return False, None

def printBoard(board):
    for i in board:
        for j in i:
            #Print row
            print(j, end = "  ")
        print("")
    print("\n")
    
if __name__ == "__main__":
    N = int(input("Enter number of Queens : "))
    result, solution = solve(N)
    if result:
        print("\nSolution Found : ")
        printBoard(solution)
    else:
        print("Solution Not Found")
```

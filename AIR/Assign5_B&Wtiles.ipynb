```python
class Problem:
    def __init__(self, data):
        self.data = data
        self.dist = self.calculate_distance()
    
    def calculate_distance(self):
        idx = -1
        dist = [0] * len(self.data)
        for i in range(len(self.data)):
            if self.data[i] == 'W':
                idx = i
            elif idx == -1:
                dist[i] = 0
            else:
                dist[i] = idx - i
        idx = -1
        for i in range(len(self.data))[::-1]:
            if self.data[i] == 'B':
                idx = i
            elif idx == -1:
                dist[i] = 0
            else:
                dist[i] = idx - i
        return dist
    
    def dist_is_nonzero(self):
        for i in self.dist:
            if i != 0:
                return True
        return False
    
    def get_tile_to_move(self, selector):
        idx = -1
        val = 0
        for i in range(len(self.data)):
            if self.data[i] == selector and 2 >= abs(self.dist[i]) > val:
                idx = i
                val = abs(self.dist[i])
        return idx
    
    def move_tile(self, idx, dist):
        while(dist < 0):
            self.data[idx], self.data[idx-1] = self.data[idx-1], self.data[idx]
            idx-=1
            dist+=1
        while(dist > 0):
            self.data[idx], self.data[idx+1] = self.data[idx+1], self.data[idx]
            idx+=1
            dist-=1

    def sort_tiles(self):
        step_cnt = 0
        total_cost = 0
        selector = 'W'
        print("\nStep : " + str(step_cnt) + "\tSelector : Initial state")
        self.print_data()
        self.print_distance()
        while(self.dist_is_nonzero()):
            step_cnt += 1
            print("\nStep : " + str(step_cnt) + "\tSelector : " + selector)
            if selector == 'W':
                idx = self.get_tile_to_move(selector)
                selector = 'B'
            else:
                idx = self.get_tile_to_move(selector)
                selector = 'W'
            if abs(self.dist[idx]) <= 2:
                self.move_tile(idx, self.dist[idx])
                cost = abs(self.dist[idx])
                self.dist = self.calculate_distance()
                self.print_data()
                self.print_distance()
                print("Cost = " + str(cost))
                total_cost += cost
        return total_cost
                
    def print_data(self):
        for i in self.data:
            print(i, end = "\t")
        print("\n")
                
    def print_distance(self):
        for i in self.dist:
            print(i, end = "\t")
        print("\n")
        
if __name__ == "__main__":
    data = ['B', 'W', 'B', 'W', 'B', 'W']
    pr = Problem(data)
    print("\n\nTotal Cost = " + str(pr.sort_tiles()))
```
